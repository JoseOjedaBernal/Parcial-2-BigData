```python
import time
import numpy as np
import pandas as pd
from pandas.io.html import read_html
import json

from bs4 import BeautifulSoup
from selenium import webdriver
from selenium.webdriver.chrome.options import Options
```

```python
SITE_URL = 'https://es.wikipedia.org/wiki/Anexo:Presidentes_de_Colombia'
```

```python
chrome_options = Options()
chrome_options.add_argument('--headless')
chrome_options.add_argument('--no-sandbox')
chrome_options.add_argument('--disable-dev-shm-usage')
prefs = {"download.default_directory" : "data/"}
chrome_options.add_experimental_option("prefs",prefs)

driver = webdriver.Chrome('/usr/bin/chromedriver',chrome_options=chrome_options)
```

```python
def make_request(browser):
    # Making the request and rendering the browser
    browser.get(SITE_URL)
    
    # Simulating vertical scrolling for handling lazy load
    check_height = browser.execute_script('return document.body.scrollHeight;')
    while True:
        browser.execute_script('window.scrollTo(0, document.body.scrollHeight);')
        time.sleep(5)
        height = browser.execute_script('return document.body.scrollHeight;')
        if height == check_height: 
            break 
        check_height = height
    
    # Getting HTML content and passing it to BeautifulSoup for scraping analysis
    return BeautifulSoup(browser.page_source, 'html.parser')
```

```python
soup = make_request(driver)
right_table = soup.find_all(class_ = 'wikitable')

A = []
B = []
C = []
D = []
E = []
F = []

for td in right_table[0].find_all('td'):
    A.append(td.text)
for td in right_table[1].find_all('td'):
    B.append(td.text)
for td in right_table[2].find_all('td'):
    C.append(td.text)
for td in right_table[3].find_all('td'):
    D.append(td.text)
for td in right_table[4].find_all('td'):
    E.append(td.text)
for td in right_table[5].find_all('td'):
    F.append(td.text)

```

```python

AA = np.array(A).reshape(-1, 7)
BB = np.array(B).reshape(-1, 6)
CC = np.array(C).reshape(-1, 6)
DD = np.array(D).reshape(-1, 6)
EE = np.array(E).reshape(-1, 6)
FF = np.array(F).reshape(-1, 6)

df = pd.DataFrame([AA][0])
df = pd.DataFrame([BB][0])
df = pd.DataFrame([CC][0])
df = pd.DataFrame([DD][0])
df = pd.DataFrame([EE][0])
df = pd.DataFrame([FF][0])
df = df.drop(df.columns[[0, 2, 5]], axis=1)
df.columns = ["Periodo", "Presidente", "Partido"]
df

```
